```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import pandas as pd
import math
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_digits

def plot_embedding(data):
    x_min, x_max = np.min(data, 0), np.max(data, 0)
    data = (data - x_min) / (x_max - x_min)
    return data

df = pd.read_csv('result/result_summarize.csv')
X_unknown = []
for img_id in ['ziyi', 'qian', 'amy', 'kiat', 'yuexin']:
    x = df[(df['img'].str.match(img_id) )].values[:,2:].reshape(-1)
    X_unknown.append(x)
X_unknown = np.array(X_unknown)

X_known = []
for img_id in ['bruce', 'cristiano', 'ellen', 'taylor', 'beyonce']:
    x = df[(df['img'].str.match(img_id) )].values[:,2:].reshape(-1)
    X_known.append(x)
X_known = np.array(X_known)
```

```python
X2 = pd.read_csv('result/ziyi_qian_amy_kiat_yuexin.csv').values[:,1:]
X = np.concatenate((X_known, X_unknown), axis=0)
colors = [0 for i in range(5)] + [1 for i in range(5)]
labels = ['SD known' for i in range(5)] + ['SD unknown' for i in range(5)]
clset = set(zip(colors, labels))
tsne_digits = TSNE(n_components=2, random_state=35, perplexity=5).fit_transform(X)
aim_data = plot_embedding(tsne_digits)

ax = plt.gca()
sc = plt.scatter(aim_data[:, 0], aim_data[:, 1], c=colors)
handles = [plt.plot([],color=sc.get_cmap()(sc.norm(c)),ls="", marker="o")[0] for c,l in clset ]
labels = [l for c,l in clset]
ax.legend(handles, labels)
plt.title("T-SNE people")
plt.show()
```
